# Targetes grogues

```python
# Unir les dades per obtenir la columna 'leagueID'
merged_data = appearences.merge(games[['gameID', 'leagueID']], on='gameID', how='left')

# Agrupar per lliga i sumar les targetes grogues
yellow_cards_per_league = merged_data.groupby('leagueID_x')['yellowCard'].sum().reset_index()

# Afegir el nom de la lliga
yellow_cards_per_league = yellow_cards_per_league.merge(leagues[['leagueID', 'name']], left_on='leagueID_x',right_on='leagueID', how='left')

# Imprimir el resultat
yellow_cards_per_league.to_csv("targetes_grogues.csv")
```

# Targetes vermelles

```python
# Unir les dades per obtenir la columna 'leagueID'
merged_data = appearences.merge(games[['gameID', 'leagueID']], on='gameID', how='left')

# Agrupar per lliga i sumar les targetes grogues
yellow_cards_per_league = merged_data.groupby('leagueID_x')['redCard'].sum().reset_index()

# Afegir el nom de la lliga
yellow_cards_per_league = yellow_cards_per_league.merge(leagues[['leagueID', 'name']], left_on='leagueID_x',right_on='leagueID', how='left')

# Imprimir el resultat
yellow_cards_per_league.to_csv("targetes_vermelles.csv")
```

# Top 10 millors equips

```python
# Assignar punts segons el resultat
def assign_points(row):
    if row['homeGoals'] > row['awayGoals']:
        return 3, 0  # Home win
    elif row['homeGoals'] < row['awayGoals']:
        return 0, 3  # Away win
    else:
        return 1, 1  # Draw

games[['points_home', 'points_away']] = games.apply(assign_points, axis=1, result_type="expand")

# Sumar els punts per cada equip
points_home = games.groupby(['homeTeamID', 'leagueID'])['points_home'].sum().reset_index()
points_away = games.groupby(['awayTeamID', 'leagueID'])['points_away'].sum().reset_index()

# Canviar els noms de les columnes per fer els merges més fàcils
points_home.rename(columns={'homeTeamID': 'teamID'}, inplace=True)
points_away.rename(columns={'awayTeamID': 'teamID'}, inplace=True)

# Combinar els punts de casa i fora
total_points = points_home.merge(points_away, on=['teamID', 'leagueID'], how='outer', suffixes=('_home', '_away')).fillna(0)
total_points['total_points'] = total_points['points_home'] + total_points['points_away']

# Afegir el nom de l'equip
total_points = total_points.merge(teams[['teamID', 'name']], on='teamID', how='left')

# Afegir el nom de la lliga
total_points = total_points.merge(leagues[['leagueID', 'name']], on='leagueID', how='left', suffixes=('_team', '_league'))

# Ordenar els equips per punts totals en ordre descendent
top_teams = total_points.sort_values(by='total_points', ascending=False).head(10)

# Coordenades aproximades de les ciutats dels equips
coordinates = {
    "Barcelona": (41.3809, 2.1228),
    "Juventus": (45.0703, 7.6869),
    "Paris Saint Germain": (48.8566, 2.3522),
    "Real Madrid": (40.4168, -3.7038),
    "Manchester City": (53.4808, -2.2426),
    "Bayern Munich": (48.1351, 11.5820),
    "Atletico Madrid": (40.4168, -3.7038),
    "Napoli": (40.8518, 14.2681),
    "Liverpool": (53.4084, -2.9916),
    "Roma": (41.9028, 12.4964)
}

# Afegir les coordenades al dataframe
top_teams['latitude'] = top_teams['name_team'].map(lambda team: coordinates.get(team, (None, None))[0])
top_teams['longitude'] = top_teams['name_team'].map(lambda team: coordinates.get(team, (None, None))[1])

# ciutats dels equips
city = {
    "Barcelona": "Barcelona",
    "Juventus": "Turin",
    "Paris Saint Germain": "Paris",
    "Real Madrid": "Madrid",
    "Manchester City": "Manchester",
    "Bayern Munich": "Munich",
    "Atletico Madrid": "Madrid",
    "Napoli": "Napoli",
    "Liverpool": "Liverpool",
    "Roma": "Roma"
}
top_teams['city'] = top_teams['name_team'].map(lambda team: city.get(team, (None, None)))
# Imprimir els 10 millors equips amb les coordenades
print(top_teams[['name_team', 'total_points', 'name_league', 'latitude', 'longitude','city']])

# Opcionalment, guardar els resultats en un fitxer CSV
top_teams.to_csv('top_teams_2014_2020.csv', index=False)
```

# 10 millors jugadors

```python
# Filtrar les dades per obtenir només els gols
goals_data = shots[shots['shotResult'] == 'Goal']

# Obtenir els màxims golejadors
top_scorers = goals_data.groupby('shooterID').size().reset_index(name='goals')

# Obtenir la suma de xGoals per jugador
xgoals_per_player = goals_data.groupby('shooterID')['xGoal'].sum().reset_index()

print(top_scorers)
# Obtenir el nombre d'assistències on el shotType sigui 'Goal' i el assisterID sigui el seu ID
assists_data = goals_data[goals_data['assisterID'].notna()]
assists_per_player = assists_data.groupby('assisterID').size().reset_index(name='assists')

# Unir les dades
top_scorers = top_scorers.merge(players[['playerID', 'name']], left_on='shooterID', right_on='playerID', how='left')
top_scorers = top_scorers.merge(xgoals_per_player, left_on='shooterID', right_on='shooterID', how='left')
top_scorers = top_scorers.merge(assists_per_player, left_on='shooterID', right_on='assisterID', how='left')
top_scorers = top_scorers[['name', 'goals', 'xGoal', 'assists']]

# Omplir NaN amb 0 per les assistències
top_scorers['assists'] = top_scorers['assists'].fillna(0).astype(int)

# Ordenar els jugadors per nombre de gols en ordre descendent i obtenir els 10 màxims golejadors
top_10_scorers = top_scorers.sort_values(by='goals', ascending=False).head(10)

# Imprimir la taula final
print(top_10_scorers)

top_10_scorers.to_csv('top_10_scorers.csv', index=False)
```
